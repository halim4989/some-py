```python
#@markdown ---
#@markdown ### <center>All torrents magnet separated by 'space'
#@markdown ---



#@markdown ---
torrent_all = "" # @param {"type":"raw"}
save_path = "/content/drive/MyDrive/mvi" # @param {type: "string"}
#@markdown ---
#@markdown ---

torrentList = torrent_all.split(' ')
print(len(torrentList))

for t in torrentList:
  info = lt.parse_magnet_uri(t)
  info.save_path = save_path
  ses.add_torrent(info)
  print(info.name)
```

```python
# total torrent count

allTorrents = [i.status() for i in ses.get_torrents()]
len(allTorrents)
```

```python
# %%time

allTorrents = [i.status() for i in ses.get_torrents()]

while 1:
  time.sleep(3)
  print("\n\n")

  allTorrents = ses.refresh_torrent_status(allTorrents)
  i=0
  for s in allTorrents:
    eta_s = ((s.total_wanted-s.total_download) / s.download_rate) if s.download_rate else 99999999
    eta = str(timedelta(seconds = eta_s))
    print(f'\n{i} {s.name}')
    i+=1

    print(f"{s.progress * 100 :7.3f}%, d:{s.download_rate/1000 :.1f}, u:{s.upload_rate/1000 :.1f}, p:{s.num_peers}, {s.state}, {s.total_wanted_done/(10**9) :.5f}/{ s.total_wanted/(10**9) :.5f},  ETA: {eta}")
```
