```python
import requests
from bs4 import BeautifulSoup

HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36',
           'Accept-Language': 'en,en-US;q=0.9'}

# key = "tattoo coloring books"

def amzsearch(wrd):
  # print(type(wrd))
  # print(wrd)
  res = requests.get(f"https://www.amazon.com/s?k={wrd}&ref=nb_sb_noss_1", headers=HEADERS)

  if res.status_code == 200:
    # print('OK')
    soup = BeautifulSoup(res.content, features="lxml")
    count = soup.select_one('.a-spacing-small.a-spacing-top-small span:nth-of-type(1)').text.split()[-3]
    return count

  else:
    print(res.status_code)
    return False


amzsearch('tattoo coloring books')
```

```python
import threading
import requests

HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36',
           'Accept-Language': 'en,en-US;q=0.9'}

def dorun(urls, i):
  # print(i, end=' ')
  for url in urls:
    requests.get(url, headers=HEADERS)
```

```python
n = 1

while True:
  print('pass ',n)
  n+=1

  threads =  [threading.Thread(target=dorun, args=(urls, i)) for i in range(99)]

  for t in threads:
      t.start()
  for t in threads:
      t.join()
```
