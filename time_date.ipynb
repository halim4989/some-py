```python
from datetime import datetime   
from pytz import common_timezones
import pytz


# UTC timezone Datetime
dt_local = datetime.now(pytz.utc)
print("UTC DateTime:", dt_local.strftime("%Y:%m:%d %H:%M:%S %Z %z"))


# Convert US/Eastern timezone to Asia/Dhaka timezone
dt_bd = dt_local.astimezone(pytz.timezone('Asia/Dhaka'))
print("Bangladesh DateTime:", dt_bd.strftime("%Y:%m:%d %H:%M:%S %Z %z"))

dt_bd.strftime("%Y%m%dT%I%M%S%P")
```

```python
dt_bd.strftime("%Y%m%d_%I%M%S%P")
```
