```python
import requests

def get_exchange_rate(from_currency, to_currency):
  # send a GET request to the API endpoint
  response = requests.get(f"https://api.exchangerate-api.com/v4/latest/{from_currency}")

  # convert the response to a dictionary
  data = response.json()

  # get the exchange rate from the dictionary
  exchange_rate = data['rates'][to_currency]

  return exchange_rate

# test the function
bdt_to_usd = get_exchange_rate('BDT', 'USD')
print(f"1 BDT = {bdt_to_usd:.2f} USD")

usd_to_bdt = get_exchange_rate('USD', 'BDT')
print(f"1 USD = {usd_to_bdt:.2f} BDT")
```

```python
import requests
from bs4 import BeautifulSoup

def get_exchange_rate():
    # Make a request to Google Search
    response = requests.get("https://www.google.com/search?q=usd+to+bdt+exchange+rate")

    # Parse the response with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the element containing the exchange rate
    exchange_rate_element = soup.find('div', {'class': 'BNeawe iBp4i AP7Wnd'})
    if exchange_rate_element:
        # Extract the exchange rate text and remove any spaces
        exchange_rate_text = exchange_rate_element.text.strip()
        return exchange_rate_text
    else:
        return "Exchange rate not found."

print(get_exchange_rate())
```
