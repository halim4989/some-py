```python
import requests, pandas as pd
from bs4 import BeautifulSoup


url = "https://www.gutenberg.org/ebooks/search/?sort_order=downloads"

# Total pages to scrap
totalPage = 5


df = pd.DataFrame()
page = 1


HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36',
           'Accept-Language': 'en,en-US;q=0.9'}

while url is not None:
  res = requests.get(url , headers=HEADERS)
  
  soup = BeautifulSoup(res.content, features="lxml")


  title = [x.text for x in soup.select(".booklink .title")]
  author = [x.text for x in soup.select(".subtitle")]
  dwn = [int(x.text[:-10]) for x in soup.select(".extra , .booklink:nth-child(10) .subtitle")]
  print(len(title))
  print(len(author))
  print(len(dwn))


  df1 = pd.DataFrame({'Title': title,
                    'Author': author,
                    'Downloads': dwn})
  df = pd.concat([df , df1])



  if soup.select_one(".links a:nth-of-type(1)").text == "Next":
    url = "https://www.gutenberg.org" + soup.select_one(".links a:nth-of-type(1)").attrs['href']
  elif soup.select_one(".links a:nth-of-type(3)").text == "Next":
    url = "https://www.gutenberg.org" + soup.select_one(".links a:nth-of-type(3)").attrs['href']
  else:
    url = None
    print("END OF PAGES")

  page += 1
  if page > totalPage:
    url = None
    print("DONE")


```

```python
df.to_excel("out.xlsx", encoding='utf8', index=False)
```
